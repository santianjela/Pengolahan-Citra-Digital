```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
foto = cv2.imread('/anjela/kampus/sem. 4/PCD/rumus.png')
```

```python
cv2.rectangle(foto, pt1 = (200, 100), pt2 = (250, 140), 
              color = (255, 0, 0), thickness = 2)
plt.imshow(foto)
```

```python
cv2.circle(foto, center = (250, 30), radius = 10, 
           color = (0, 0, 255), thickness = 2)
plt.imshow(foto)
```

```python
cv2.putText(foto, text = "the Wall of Love", 
            org = (20, 80),
            fontFace = cv2.FONT_HERSHEY_DUPLEX, 
            fontScale = 1, 
            color = (0, 255, 0), 
            thickness = 1, 
            lineType = cv2.LINE_AA)
plt.imshow(foto)
```
